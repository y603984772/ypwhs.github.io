````python
import re
import os
import shutil
````

````python
path = '_posts/'
posts = os.listdir(path)
````

````python
path = '_posts/'
posts = os.listdir(path)

for i, p in enumerate(posts):
    if p == '.DS_Store':
        continue
    with open(path + p, 'r') as f:
        content = f.read()
        title = re.findall('title: (.+)', content)[0]
        print i, title, p
        shutil.move(path+p, path+title+'.md')
````

````python
posts = os.listdir(path)

for i, p in enumerate(posts):
    with open(path + p, 'r') as f:
        content = f.read()
        images = re.findall('\[\!\[.*?\]\(http://.+/(.+?\..+?)\)', content)
        images2 = re.findall('\[\!\[.*?\]\(http://.+/.+?\)', content)
        
        print p
        for a, b in zip(images, images2):
            print a, b
            content = content.replace(b, '[![](/images/%s)](/images/%s)' % (a,a) )
    
    with open(path + p, 'w') as f:
        f.write(content)
````

````python
posts = os.listdir(path)

for i, p in enumerate(posts):
    with open(path + p, 'r') as f:
        content = f.read()

    content = content.replace('tags:\n', '')
    content = content.replace('categories', 'tags')
    
    with open(path + p, 'w') as f:
        f.write(content)
````

````python
posts = os.listdir(path)

for i, p in enumerate(posts):
    with open(path + p, 'r') as f:
        content = f.read()
    
    lang = re.findall('<pre class="lang:(\w+) .+?>', content)
    html = re.findall('<pre.+?>', content)
    for a, b in zip(lang, html):
        print a, b
        if a != 'default':
            if a == 'shell':
                a = 'sh'
            content = content.replace(b, '```%s\n' % a)
        else:
            content = content.replace(b, '```\n')
    content = content.replace('</pre>', '\n```')
    
    with open(path + p, 'w') as f:
        f.write(content)
````

````python
posts = os.listdir(path)

for i, p in enumerate(posts):
    with open(path + p, 'r') as f:
        content = f.read()

    content = content.replace('&lt;', '<')
    content = content.replace('&gt;', '>')
    content = content.replace('&quot;', '"')
    content = content.replace('&amp;', '&')
    content = content.replace('&nbsp;', ' ')
    
    
    with open(path + p, 'w') as f:
        f.write(content)
````
